```python
movie = pd.read_csv('Movie-Ratings.csv')
```

```python
movie.head()
```

```python
movie.columns = ['Film','Genre','CriticRating','AudienceRating','BudgetMillions',\
                'Year']
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style("darkgrid")
f,axes = plt.subplots(2,2,figsize=(15,15))
k1 = sns.kdeplot(movie.BudgetMillions,movie.AudienceRating,ax=axes[0,0])
k2 = sns.kdeplot(movie.BudgetMillions,movie.CriticRating,ax=axes[0,1])
z = sns.violinplot(data=movie[movie.Genre=='Drama'],x='Year',y='CriticRating',ax=axes[1,0])
k4 = sns.kdeplot(movie.CriticRating,movie.AudienceRating,shade=True,shade_lowest=False)
k4b = sns.kdeplot(movie.CriticRating,movie.AudienceRating,cmap='Reds',ax=axes[1,1])
axes[1,1].hist(movie.CriticRating,bins=15)
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
sns.set_style("dark",{"axes.facecolor":"black"})
f,axes = plt.subplots(2,2,figsize=(15,15))
k1 = sns.kdeplot(movie.BudgetMillions,movie.AudienceRating,\
                 shade=True,shade_lowest=True,cmap='inferno',ax=axes[0,0])
k2 = sns.kdeplot(movie.BudgetMillions,movie.CriticRating,\
                 shade=True,shade_lowest=True,cmap='seismic_r',ax=axes[0,1])
z = sns.violinplot(data=movie[movie.Genre=='Drama'],x='Year',y='CriticRating',ax=axes[1,0])
k4 = sns.kdeplot(movie.CriticRating,movie.AudienceRating,cmap='Blues_r',shade=True,shade_lowest=False)
k4b = sns.kdeplot(movie.CriticRating,movie.AudienceRating,cmap='gist_gray_r',ax=axes[1,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```
